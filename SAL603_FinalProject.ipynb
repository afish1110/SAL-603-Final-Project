```python
##displays all outputs of each cell
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##cy young file data taken from fangraphs
cy_young_pitch_mix = pd.read_csv('data/CYYoungPitchMix.csv')

##usage is the percent that each pitch was thrown
cy_young_pitch_mix['Usage'] = round((cy_young_pitch_mix.num_pitches / cy_young_pitch_mix.total_pitches) * 100, 2)

cy_young_pitch_mix
```

```python
##df with pitch_type counted for histogram of num pitches in arsenal for cy young winners
pitch_arsenal = cy_young_pitch_mix[['Player', 'Season', 'pitch_type']].groupby(['Player', 'Season']).count()
fig = plt.figure()

##histogram no legend
ax = sns.histplot(data = pitch_arsenal, legend = False, binrange = (3, 6.5))
ax.set_xlabel('Number of Pitches')
ax.set_ylabel('Count')
ax.set_title('Number of Pitches in Arsenal of Cy Young Winners 2010-19')

fig.savefig('CyYoungArsenalCount.jpg')
```

```python
fig = plt.figure()

##boxplot shows average velocity by pitch type
ax = sns.boxplot(x = 'pitch_type', y = 'Velocity', data = cy_young_pitch_mix, palette = 'colorblind')
ax.set_xlabel('Pitch Type')
ax.set_ylabel('Average Velocity')
ax.set_title('Average Velocity by Pitch Type - Cy Young Winners 2010-19')

fig.savefig('CyYoungPitchVelo.jpg')
```

```python
fig, axes = plt.subplots(2, 1)

##bar chart of usage by pitch type
sns.barplot(x = 'pitch_type', y = 'Usage', ax = axes[0], data = cy_young_pitch_mix, color = 'green', errorbar = None)
axes[0].set_xlabel('')
axes[0].set_ylabel('Average Percent Thrown')

##bar chart opponenet avg by pitch type
sns.barplot(x = 'pitch_type', y = 'AVG', ax = axes[1], data = cy_young_pitch_mix, color = 'red', errorbar = None)
axes[1].set_xlabel('Pitch Type')
axes[1].set_ylabel('Opponent AVG')

fig.suptitle('Pitch Usage and Opponent Average - Cy Young Winners 2010-19')

fig.savefig('CyYoungUsage&AVG.jpg')
```
